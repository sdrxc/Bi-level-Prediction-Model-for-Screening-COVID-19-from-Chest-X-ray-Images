```python
import gzip
import numpy as np
import pandas as pd
from time import time
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
import tensorflow as tf
import keras
import keras.layers as layers
from keras.models import Sequential
from keras.preprocessing.image import ImageDataGenerator
from keras.utils.np_utils import to_categorical
from keras.callbacks import TensorBoard

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import pickle
model_l1 = pickle.load(open('/content/drive/My Drive/CART L1.sav', 'rb'))
model_l2 = pickle.load(open('/content/drive/My Drive/KNN L1.sav', 'rb'))
model_l3 = pickle.load(open('/content/drive/My Drive/LR L1.sav', 'rb'))
model_l4 = pickle.load(open('/content/drive/My Drive/RF L1.sav', 'rb'))
model_l5 = pickle.load(open('/content/drive/My Drive/XGB L1.sav', 'rb'))
```

```python
df=pd.read_csv('/content/drive/My Drive/level1 test_v2.csv')
df.columns
df=df.drop(columns=['Unnamed: 0'])

X1=df.iloc[:,0:25088]
y1=df['Label']
df
```

```python
X1
```

```python
from sklearn.metrics import precision_recall_curve, roc_curve, roc_auc_score

y1_l1_pred= model_l1.predict(X1)
print(metrics.classification_report(y1,y1_l1_pred))

cart_precision, cart_recall, _ = precision_recall_curve(y1,y1_l1_pred)
plt.plot(cart_recall, cart_precision, marker='.', label='CART')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()

y1_l2_pred= model_l2.predict(X1)
print(metrics.classification_report(y1,y1_l2_pred))


knn_precision, knn_recall, _ = precision_recall_curve(y1,y1_l2_pred)
plt.plot(knn_recall, knn_precision, marker='.', label='kNN')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()

y1_l3_pred= model_l3.predict(X1)
print(metrics.classification_report(y1,y1_l3_pred))


mod_probs = model_l3.predict_proba(X1)
# keep probabilities for the positive outcome only
mod_probs = mod_probs[:, 1]
ns_probs = [0 for _ in range(len(X1))]
# calculate scores
# ns_auc = roc_auc_score(y1, ns_probs)
# mod_auc = roc_auc_score(y1, lr_probs)
# # summarize scores
# print('No Skill: ROC AUC=%.3f' % (ns_auc))
# print('XGB: ROC AUC=%.3f' % (mod_auc))
# calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(y1, ns_probs)
mod_fpr, mod_tpr, _ = roc_curve(y1, mod_probs)
# plot the roc curve for the model
plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Discrimination Line')
plt.plot(mod_fpr, mod_tpr, marker='.', label='LR + XGB')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.savefig('/content/drive/My Drive/matplot.jpg')
plt.show()

mod_auc = roc_auc_score(y1, mod_probs)
mod_auc
```

```python
lr_precision, lr_recall, _ = precision_recall_curve(y1,y1_l3_pred)
plt.plot(lr_recall, lr_precision, marker='.', label='LR')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()


y1_l4_pred= model_l4.predict(X1)
print(metrics.classification_report(y1,y1_l4_pred))


rf_precision, rf_recall, _ = precision_recall_curve(y1,y1_l4_pred)
plt.plot(rf_recall, rf_precision, marker='.', label='RF')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()


y1_l5_pred= model_l5.predict(X1)
print(metrics.classification_report(y1,y1_l5_pred))
```
